```python
import pandas as pd
import numpy  as np
```

```python
df = pd.read_csv('./flat-training.csv')

display(df.dtypes)
display(df.head())
```

```python
num_cols = df.select_dtypes(include='number').columns.tolist()
cat_cols = df.select_dtypes(exclude='number').columns.tolist()
```

```python
cols_na = df.columns[df.isnull().sum() > 0].tolist() # 결측치 열 추출

df = df.drop(columns=cols_na) # 결측치 열 제거

df.head()
```

```python
df.dtypes
```

```python
with np.load('pbmc_x.npz') as data:
    X = data['arr_0']
```

```python
unique, counts = np.unique(X, return_counts=True)

print(unique)
print(counts)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

# 범주형 데이터를 Label Encoding 진행
for col in cat_cols:
    le = le.fit(df[col])
    df[col] = le.transform(df[col])

df[cat_cols].head()
```
